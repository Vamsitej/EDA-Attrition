### Doing a check on overtime variable having any linkage with attrition

```python
#Plotting employee distribution by OverTime, with distinct Attrition classes
sns.axes_style('whitegrid')
sns.factorplot("OverTime", data=df, aspect=1, kind='count', hue='Attrition', palette=['C1', 'C0']).set_ylabels('Number of Employees')
```

```python
ot=pd.crosstab(df.Attrition,columns=df.OverTime)
ot
```

```python
rowtotal_ot=ot.sum(axis=1)
rowtotal_ot
```

```python
percentage_ot=round(ot.div(rowtotal_ot,axis=0)*100,2)
percentage_ot

#Higher proportion of people who left the organization worked overtime 
#23.44% of people with attiriton 0 wokred overtime
#53.6% of people with attrition 1 worked overtime
#Below data shows that overtime is strongly correlated with attrition
```
